```python
print("Cities Selection Based on Low Crime")
print("To figure out cities with low crime, I first cleaned up the data then got the descriptive statistics.")

import pandas as pd
Crimedata = pd.read_csv("C:\\Users\\tangw\\Desktop\\2021 Fall\\HUDK 4050\\Data\\Crime_2015.csv")
CrimeClean = Crimedata.dropna()
CrimeClean
```

```python
CrimeClean.describe()
```

```python
print("I then chose the range for all variables to be <25% to select cities with low crime rates overall.")

CrimeMask = (CrimeClean['Murder'] < 2.225000) & (CrimeClean['Rape'] <29.000000) & (CrimeClean['Robbery'] <37.025000) & (CrimeClean['AggravatedAssault'] <144.300000) & (CrimeClean['MotorVehicleTheft'] <98.500000)
CrimeClean[CrimeMask]
FewCrime = CrimeClean[CrimeMask]
FewCrime
```

```python
print("As the result shown, 11 cities are selected for low crime rates, Which are:")
print("Appleton, Bangor, Bloomsburg, Daphne, Eau Claire, La Crosse, Montgomery, Portland, Provo, Wausau")
```
